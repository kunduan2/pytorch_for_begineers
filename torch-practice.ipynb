```python
#  setting the random seed for numpy
print("In Numpy: \n")
np.random.seed(42)
# matrix of random numbers
print(np.random.rand(3,2)) 
print(np.random.randn(3,3),"\n") # also print random num from -ve axis and follows normal dist.


#### Try same in torch
print("In torch: \n")
torch.manual_seed(42)
print(torch.rand(3,3))

```

```python
torch.manual_seed(42)
a = torch.randint(0,10, size = (3,3))
b = torch.randint(1,10, size = (3,3))
print(a,"\n",b, "\n")
# matrix addition
# print(torch.add(a,b), '\n')
# print(a+b, '\n')

# matrix subtraction
# print(torch.sub(a,b), '\n')

# matrix multiplication
# print(torch.mm(a,b), '\n')

# elementwise multiplication
print(a*b, '\n')

# # matrix division
# print(torch.div(a,b))
```

```python
# concatenating vertically (adding row-wise, column number=fixed)
torch.cat((a,b)) 
```

```python
# concatenating horizontally (adding column-wise, row number=fixed)
torch.cat((a,b), dim=0)
```

```python
# initializing a numpy array
a = np.array([[1,2],[3,4]])
print(a, '\n')

# converting the numpy array to tensor
tensor = torch.from_numpy(a)
print(tensor)
```

```python
# calculate gradient 
a = torch.ones((2,2), requires_grad=True)
b = a + 5
c = b.mean()
c = c**2
# c = c + 56

print(b,c)

# back propagating
c.backward()

# computing gradients
print(a.grad)
```
